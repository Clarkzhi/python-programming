```python
import pandas as pd
from pandas.io.json import json_normalize
import json
import dateutil
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
NBA = pd.read_json('NBA.json')
```

```python
NBA.head()
```

```python
games = json_normalize(NBA['api']['games'])
```

```python
games
```

```python
games.info()
```

```python
games.describe()
```

```python
games.columns
```

```python
#drop some columns
games.drop(columns=['vTeam.teamId', 'clock', 'currentPeriod', 'seasonStage', 'statusShortGame', 'hTeam.teamId', 'vTeam.logo', 'hTeam.logo', 'gameId', 'halftime', 'statusGame' ], inplace=True)
```

```python
games.head(20)
```

```python
#add a new column name is 'results'
games['results'] = games['hTeam.score.points'] >= games['vTeam.score.points']
```

```python
games.head(30)
```

```python
def result_name(val):
    result = ['win', 'lose']
    try:
        return result[val - 1]
    except:
        return val
    
```

```python
games['results'] = games['results'].apply(result_name)
```

```python
games.head(20)
```

```python
games['gameDuration'].value_counts().sort_values(ascending = False).plot(figsize=(20, 5))
```

```python
#how many games in each country and which leagues are they.
games['league'].value_counts()
```

```python
games.groupby('country')['league'].value_counts().unstack().plot.bar(figsize = (20,20))
```

```python
games['country'].value_counts()
```

```python
#longgest game in Canada
games[games['country'] == 'Canada']['gameDuration'].max()
```

```python
games[(games['country'] == 'Canada') & (games['gameDuration'] == '2:38')][['startTimeUTC', 'arena', 'hTeam.nickName', 'hTeam.score.points', 'vTeam.nickName', 'vTeam.score.points', 'results']]
```

```python
#shortest game in China
games[games['country'] == 'China']['gameDuration'].min()
```

```python
games[(games['country'] == 'China') & (games['gameDuration'] == '2:28')][['startTimeUTC', 'arena', 'hTeam.nickName', 'hTeam.score.points', 'vTeam.nickName', 'vTeam.score.points', 'results']]
```

```python
#games in Canada
games[games['country'] == 'Canada'][['startTimeUTC','arena', 'vTeam.nickName', 'vTeam.score.points', 'hTeam.nickName','hTeam.score.points', 'gameDuration', 'results']].sort_values(by = 'startTimeUTC')
```

```python
#how many games win in canada and how many lose
games[games['country'] == 'Canada'].groupby('country')['results'].value_counts().unstack()
```

```python
#How many satndard games win at host 
games[(games['results'] == 'win') & (games['league'] == 'standard')].groupby(['country','city', 'arena', 'hTeam.nickName'])[['results']].count().sort_values(by = 'results', ascending = False)
```
