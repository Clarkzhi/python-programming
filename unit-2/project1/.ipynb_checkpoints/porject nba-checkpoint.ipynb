```python
import pandas as pd
from pandas.io.json import json_normalize
import json
import dateutil
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
NBA = pd.read_json('NBA.json')
```

```python
NBA.head()
```

```python
games = json_normalize(NBA['api']['games'])
```

```python
games
```

```python
games.info()
```

```python
games.describe()
```

```python
games.columns
```

```python
#drop some columns
games.drop(columns=['vTeam.teamId', 'hTeam.teamId', 'vTeam.logo', 'hTeam.logo', 'gameId', 'halftime', 'statusGame' ], inplace=True)
```

```python
games.head()
```

```python
games['results'] = games['hTeam.score.points'] >= games['vTeam.score.points']
```

```python
games.head()
```

```python
def result_name(val):
    result = ['win', 'lose']
    try:
        return result[val - 1]
    except:
        return val
    
```

```python
games['results'] = games['results'].apply(result_name)
```

```python
games.head()
```

```python
#sort by starttimeutc
games.groupby('startTimeUTC').first()
```

```python
games['city'].value_counts().plot.bar()
```

```python
games[games['results'] == 'win'].groupby('seasonYear')[['results']].count()
```

```python
#longgest game in Canada
games[games['country'] == 'Canada'].groupby('seasonYear')[['gameDuration']].max()
```

```python
games[games['country'] == 'Canada'].groupby('seasonYear')[['gameDuration']].min()
```

```python
#games in Canada
games[games['country'] == 'Canada'].groupby(['startTimeUTC','arena', 'vTeam.nickName', 'vTeam.score.points', 'hTeam.nickName','hTeam.score.points','results'])[['gameDuration']].count() 
```

```python
#How many wins in host
games[games['results'] == 'win'].groupby(['country','city', 'arena'])[['results']].count()
```
